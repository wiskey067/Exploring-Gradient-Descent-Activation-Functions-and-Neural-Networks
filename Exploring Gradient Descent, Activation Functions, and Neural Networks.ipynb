```python
#PART1:GRADIENT DESCENT ON A CUSTOM FUNCTION
```

```python

import numpy as np

# Defining the function
def f(x, y):
    return np.sin(x) + y**2 + x*y + 0.5 * x**2

# Partial derivatives
def grad_x_f(x, y):
    return np.cos(x) + y + x

def grad_y_f(x, y):
    return 2 * y + x
```

```python
# Gradient descent function
def gradient_descent(x_init, y_init, step_size=0.1, steps=15):
    x, y = x_init, y_init
    for i in range(steps):
        fx = f(x, y)
        print(f"Step {i+1}: x = {x:.4f}, y = {y:.4f}, f(x, y) = {fx:.4f}")
        
        # Update rules
        x -= step_size * grad_x_f(x, y)
        y -= step_size * grad_y_f(x, y)
```

```python
print("===== Using step size = 0.1 =====")
gradient_descent(2.0, -1.0, step_size=0.1)

print("\n===== Using step size = 0.01 =====")
gradient_descent(2.0, -1.0, step_size=0.01)
```

# PART 2:Activation Functions and Gradient Checking

```python

import matplotlib.pyplot as plt

# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Tanh
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# ReLU
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)
```

```python
x_vals = np.linspace(-5, 5, 100)

plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(3, 2, 1)
plt.plot(x_vals, sigmoid(x_vals), label="Sigmoid")
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(x_vals, sigmoid_derivative(x_vals), label="Sigmoid Derivative", color='orange')
plt.title("Sigmoid Derivative")
plt.grid(True)

# Tanh
plt.subplot(3, 2, 3)
plt.plot(x_vals, tanh(x_vals), label="Tanh")
plt.title("Tanh")
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(x_vals, tanh_derivative(x_vals), label="Tanh Derivative", color='orange')
plt.title("Tanh Derivative")
plt.grid(True)

# ReLU
plt.subplot(3, 2, 5)
plt.plot(x_vals, relu(x_vals), label="ReLU")
plt.title("ReLU")
plt.grid(True)

plt.subplot(3, 2, 6)
plt.plot(x_vals, relu_derivative(x_vals), label="ReLU Derivative", color='orange')
plt.title("ReLU Derivative")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def gradient_check(f, f_prime, x, epsilon=1e-5):
    numerical_grad = (f(x + epsilon) - f(x - epsilon)) / (2 * epsilon)
    analytical_grad = f_prime(x)
    print(f"x = {x}")
    print(f"Numerical Gradient = {numerical_grad}")
    print(f"Analytical Gradient = {analytical_grad}")
    print(f"Difference = {np.abs(numerical_grad - analytical_grad)}")
    print("-" * 40)
```

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return 1.0 if x > 0 else 0.0

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def gradient_check(f, f_prime, x, epsilon=1e-5):
    numerical_grad = (f(x + epsilon) - f(x - epsilon)) / (2 * epsilon)
    analytical_grad = f_prime(x)
    print(f"x = {x}")
    print(f"Numerical Gradient = {numerical_grad}")
    print(f"Analytical Gradient = {analytical_grad}")
    print(f"Difference = {abs(numerical_grad - analytical_grad)}\n")

# ReLU checks
print("Checking ReLU:")
for val in [-2.0, -1.0, 0.0, 1.0, 2.0]:
    gradient_check(relu, relu_derivative, val)

# tanh checks
print("Checking tanh:")
for val in [-2.0, -1.0, 0.0, 1.0, 2.0]:
    gradient_check(tanh, tanh_derivative, val)
```

```python
PART 3 Training Neural Networks on the digits dataset using scikit-learn.
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
digits = load_digits()
X, y = digits.data, digits.target

print("Dataset shape:", X.shape)
print("Label shape:", y.shape)
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python

results = {}

for solver in ['sgd', 'adam', 'lbfgs']:
    print(f"\nTraining MLPClassifier with solver = '{solver}'")
    
    # Create the model
    model = MLPClassifier(hidden_layer_sizes=(64,), solver=solver, max_iter=1000, random_state=42)
    
    # Time the training
    start_time = time.time()
    model.fit(X_train, y_train)
    duration = time.time() - start_time
    
    # Evaluate model
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    # Print results
    print(f"Training Accuracy: {train_score:.4f}")
    print(f"Test Accuracy:     {test_score:.4f}")
    print(f"Time Taken:        {duration:.2f} seconds")
    
    # Save results
    results[solver] = {
        'model': model,
        'train_score': train_score,
        'test_score': test_score,
        'duration': duration
    }
```

```python
plt.figure(figsize=(10,5))

for solver in ['sgd', 'adam']:
    model = results[solver]['model']
    plt.plot(model.loss_curve_, label=f'{solver} loss')

plt.title("Loss Curves")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

summary = pd.DataFrame({
    'Solver': list(results.keys()),
    'Train Accuracy': [results[s]['train_score'] for s in results],
    'Test Accuracy': [results[s]['test_score'] for s in results],
    'Training Time (s)': [results[s]['duration'] for s in results]
})

summary
```

```python
# Part 4: Training and Comparing Neural Network Optimizers
```

```python
import time
from sklearn.neural_network import MLPClassifier

results = {}

for solver in ['sgd', 'adam', 'lbfgs']:
    print(f"\nTraining MLPClassifier with solver = '{solver}'")
    
    model = MLPClassifier(hidden_layer_sizes=(64,), solver=solver, max_iter=1000, random_state=42)
    
    start_time = time.time()
    model.fit(X_train, y_train)
    duration = time.time() - start_time
    
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    print(f"Training Accuracy: {train_score:.4f}")
    print(f"Test Accuracy:     {test_score:.4f}")
    print(f"Time Taken:        {duration:.2f} seconds")
    
    results[solver] = {
        'model': model,
        'train_score': train_score,
        'test_score': test_score,
        'duration': duration
    }
```

```python
import matplotlib.pyplot as plt

solvers = list(results.keys())
train_scores = [results[sol]['train_score'] for sol in solvers]
test_scores = [results[sol]['test_score'] for sol in solvers]
times = [results[sol]['duration'] for sol in solvers]

x = range(len(solvers))

plt.figure(figsize=(12, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.bar(x, train_scores, width=0.3, label='Train Accuracy', align='center')
plt.bar([i + 0.3 for i in x], test_scores, width=0.3, label='Test Accuracy', align='center')
plt.xticks([i + 0.15 for i in x], solvers)
plt.ylim(0.9, 1.01)
plt.title('Training vs Test Accuracy by Solver')
plt.legend()

# Time Plot
plt.subplot(1, 2, 2)
plt.bar(solvers, times, color='orange')
plt.title('Training Time by Solver')
plt.ylabel('Time (seconds)')

plt.tight_layout()
plt.show()
```

```python
#Part 5: Plotting Loss Curves and Final Network Evaluation
```

```python
plt.figure(figsize=(10, 6))

for solver in ['sgd', 'adam']:
    model = results[solver]['model']
    plt.plot(model.loss_curve_, label=f'{solver.upper()}')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve During Training')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Confusion Matrix & Classification Report for Best Model
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# Find best model by test accuracy
best_solver = max(results, key=lambda k: results[k]['test_score'])
best_model = results[best_solver]['model']
y_pred = best_model.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=digits.target_names)
disp.plot(cmap='Blues')
plt.title(f'Confusion Matrix ({best_solver.upper()} Solver)')
plt.show()

# Classification report
print(f"\nClassification Report for Best Model ({best_solver.upper()} Solver):\n")
print(classification_report(y_test, y_pred, digits=4))
```
